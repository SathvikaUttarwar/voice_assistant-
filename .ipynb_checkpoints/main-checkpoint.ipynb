```python
import sys
sys.path.append(r'C:\Users\sathv\voice assistant\selenium_web.ipynb')
sys.path.append(r'C:\Users\sathv\voice assistant\youtube.ipynb')
sys.path.append(r'C:\Users\sathv\voice assistant\news.ipynb')
sys.path.append(r'C:\Users\sathv\voice assistant\joke.ipynb')
sys.path.append(r'C:\Users\sathv\voice assistant\weather.ipynb')
from selenium_web import *
from youtube import *
from news import *
from joke import *
from weather import *

import pyttsx3 as p
import speech_recognition as sr
import randfacts
import datetime 

engine = p.init()

rate = engine.getProperty('rate')
engine.setProperty('rate', 150)
voices = engine.getProperty('voices')
engine.setProperty('voice', voices[1].id)

def speak(text):
    engine.say(text)
    engine.runAndWait()

def wishme():
    hour = int(datetime.datetime.now().hour)
    if hour >= 0 and hour < 12:
        return "morning"
    elif hour >= 12 and hour < 16:
        return "afternoon"
    else:
        return "evening"

today_date = datetime.datetime.now() 
r = sr.Recognizer()

speak("Hello, good " + wishme() + " I am your voice assistant. How are you?")

def listen_command():
    with sr.Microphone() as source:
        r.energy_threshold = 10000
        r.adjust_for_ambient_noise(source, 1.2)
        print("Listening...")
        audio = r.listen(source)
        try:
            return r.recognize_google(audio).lower()
        except sr.UnknownValueError:
            speak("Sorry, I didn't catch that. Could you repeat?")
            return listen_command()

text = listen_command()

if "what" in text and "about" in text and "you" in text:
    speak("I am having a good day")
speak("What can I do for you?")

while True:
    text2 = listen_command()
    
    if "information" in text2:
        print("You need information about what topic?")
        speak("You need information about what topic?")
        infor = listen_command()
        print("Searching {} on Wikipedia".format(infor))
        speak("Searching {} on Wikipedia".format(infor))
        assist = Infow()
        assist.get_info(infor)

    elif "play" in text2 and "video" in text2:
        speak("Okay! Which video do you want to play?")
        vid = listen_command()
        print("Searching {} on YouTube".format(vid))
        speak("Searching {} on YouTube".format(vid))
        assist = music()
        assist.play(vid)

    elif "date" in text2 and "time" in text2:
        speak("Okay, sure")
        current_time = "Today is " + today_date.strftime("%d") + " of " + today_date.strftime("%B") + " and it's currently " + today_date.strftime("%I") + ":" + today_date.strftime("%M") + " minutes and " + today_date.strftime("%S") + " seconds."
        print(current_time)
        speak(current_time)

    elif "news" in text2:
        print("Sure, I will read news for you")
        speak("Sure, I will read news for you")
        arr = news()
        for i in range(len(arr)):
            print(arr[i])
            speak(arr[i])

    elif "joke" in text2 or "jokes" in text2:
        speak("Sure, I can tell you jokes")
        print("Joke:")
        arr = joke()
        print(arr[0])
        speak(arr[0])
        print(arr[1])
        speak(arr[1])

    elif "weather" in text2:
        speak("Sure")
        print(temp())
        print(des())
        speak(temp())
        speak(des())

    elif "fact" in text2 or "facts" in text2:
        speak("Sure")
        x = randfacts.get_fact()
        print(x)
        speak("Did you know that, " + x)

    elif "stop" in text2 or "exit" in text2:
        print("Goodbye!")
        speak("Goodbye!")
        break

    else:
        speak("Sorry, I didn't understand that. Can you please repeat?")








    
```
